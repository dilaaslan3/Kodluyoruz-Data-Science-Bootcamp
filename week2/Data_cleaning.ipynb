```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('bestseller_books.csv')
df2 = pd.read_csv('computer_books.csv')
df3 = pd.read_csv('engineering_books.csv')
```

```python
df_final = df2.append(df3, ignore_index = True)
df_final = df_final.append(df, ignore_index = True)
df_final
```

```python
df_final.drop_duplicates(inplace=True)
df_final.shape
```

```python
df_final['Rating'] = df_final['Rating'].apply(lambda x: x.split()[-1:][0])
df_final['Rating'] =df_final["Rating"].str.replace(',', '.')
df_final['User Rates'] =df_final["User Rates"].str.replace('.', '')
df_final
```

```python
df_final.describe()
```

```python
df_final.drop(df_final[df_final["User Rates"]=="None"].index, inplace=True)
df_final.drop(df_final[df_final["Price"] == '0'].index, inplace=True)
df_final.shape
```

```python
df_final['Price'] = df_final['Price'].apply(lambda x: x.split('\xa0')[0])
df_final["Price"] = df_final["Price"].str.replace(',', '.')
df_final
```

```python
df_final.Rating = df_final.Rating.astype(float)
df_final["User Rates"] = df_final["User Rates"].astype(int)
df_final.drop(df_final[df_final["Price"].str.len() == 8].index, inplace=True)
df_final
```

```python
df_final.Price = df_final.Price.astype(float)
```

```python
df_final.dtypes
```

```python
df_final.drop(df_final[df_final["Book Name"] == "unknown-product"].index, inplace=True)
```

```python
df_final.to_csv('biggest_ecommerce_books_data.csv', index=False, encoding='utf-8-sig')
```
